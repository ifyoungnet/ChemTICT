## Call the best model and prediction

```python
# import related packages
import pandas as pd
import joblib
```

```python
# Load the best model 
model = joblib.load('AtomPairfinal_model.pkl')

# Read the input CSV named "Outputs of Demo1.csv"
data = pd.read_csv(r"Outputs of Demo1.csv")
X = data.drop(columns=['SMILES'])

# Use the best model to  predicte the dataset
predictions = model.predict(X)

# Save your the prediction values into a CSV named "Outputs of Demo2.csv"
data['Prediction'] = predictions
output_path = r"Outputs of Demo2.csv"
data.to_csv(output_path, index=False)
```
